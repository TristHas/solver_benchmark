### First pass on solvers, unoptimized, GH200

```python
import pandas as pd
import hvplot.pandas  # noqa
import panel as pn
pn.extension()

df = pd.read_csv("benchmark_eigh_results.csv")

Ns = sorted(df.N.unique())
Bs = sorted(df.B.unique())
Ys = ["time_ms", "time_per_matrix_ms", "throughput_mat_per_s"]

wN  = pn.widgets.Select(name="N", options=Ns, value=Ns[0])
wB  = pn.widgets.Select(name="B", options=Bs, value=Bs[0])
wY  = pn.widgets.Select(name="Y", options=Ys, value="time_per_matrix_ms")

def p1(N, B, Y):
    d = df[(df.N == N) & (df.B == B)]
    return d.hvplot.bar(x="backend", y=Y, rot=45, height=350, width=500, shared_axes=False)

wB2 = pn.widgets.Select(name="B", options=Bs, value=Bs[-1])
wY2 = pn.widgets.Select(name="Y", options=Ys, value=wY.value)

def p2(B, Y):
    d = df[df.B == B]
    return d.hvplot(x="N", y=Y, by="backend", height=350, shared_axes=False)

wN3 = pn.widgets.Select(name="N", options=Ns, value=Ns[-1])
wY3 = pn.widgets.Select(name="Y", options=Ys, value=wY.value)

def p3(N, Y):
    d = df[df.N == N]
    return d.hvplot(x="B", y=Y, by="backend", height=350, shared_axes=False)

layout = pn.Row(
    pn.Column(wN, wB, wY,  pn.bind(p1, wN, wB, wY)),
    pn.layout.Divider(),
    pn.Column(wB2, wY2,    pn.bind(p2, wB2, wY2)),
    pn.layout.Divider(),
    pn.Column(wN3, wY3,    pn.bind(p3, wN3, wY3)),
)

layout  # <- IMPORTANT: last line in the cell (don’t use .servable() in a normal notebook cell)
```
